```python
# Install the correct MCP packages
import subprocess
import sys

# Install the official MCP Python SDK
packages_to_install = [
    "mcp",  # The main MCP package
    "anthropic",  # For Anthropic API integration
]

for package in packages_to_install:
    try:
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])
        print(f"✓ Successfully installed {package}")
    except subprocess.CalledProcessError as e:
        print(f"✗ Failed to install {package}: {e}")

# Now import the MCP modules
import asyncio
import json
from pathlib import Path

try:
    # Import MCP components (adjust imports based on actual package structure)
    import mcp
    from mcp import ClientSession, StdioServerParameters
    from mcp.client.stdio import stdio_client
    print("✓ MCP packages imported successfully")
except ImportError as e:
    print(f"Import error: {e}")
    print("Using fallback demo implementation...")

# Your existing SimpleMCPDemo class and demo code works perfectly as a fallback
class SimpleMCPDemo:
    def __init__(self):
        self.tools = {
            "read_file": "Read contents of a file",
            "list_files": "List files in a directory", 
            "calculate": "Perform basic calculations"
        }
        self.resources = {}
    
    def list_tools(self):
        """Simulate MCP list_tools"""
        return [{"name": name, "description": desc} for name, desc in self.tools.items()]
    
    def call_tool(self, tool_name, **kwargs):
        """Simulate MCP tool calling"""
        if tool_name == "read_file":
            filepath = kwargs.get("filepath", "")
            try:
                with open(filepath, 'r') as f:
                    return {"success": True, "content": f.read()[:500]}
            except Exception as e:
                return {"success": False, "error": str(e)}
        
        elif tool_name == "list_files":
            directory = kwargs.get("directory", ".")
            try:
                files = [str(p) for p in Path(directory).iterdir() if p.is_file()]
                return {"success": True, "files": files[:10]}
            except Exception as e:
                return {"success": False, "error": str(e)}
        
        elif tool_name == "calculate":
            expression = kwargs.get("expression", "")
            try:
                allowed_chars = set('0123456789+-*/.() ')
                if all(c in allowed_chars for c in expression):
                    result = eval(expression)
                    return {"success": True, "result": result}
                else:
                    return {"success": False, "error": "Invalid expression"}
            except Exception as e:
                return {"success": False, "error": str(e)}
        
        return {"success": False, "error": f"Unknown tool: {tool_name}"}

# Run your existing demo
async def mcp_notebook_demo():
    mcp_demo = SimpleMCPDemo()
    
    print("=== MCP Demo in Jupyter Notebook ===\n")
    
    # List available tools
    print("Available Tools:")
    tools = mcp_demo.list_tools()
    for tool in tools:
        print(f"  • {tool['name']}: {tool['description']}")
    
    print("\n" + "="*50 + "\n")
    
    # Demo: List files, calculate, and read file
    result = mcp_demo.call_tool("list_files", directory=".")
    if result["success"]:
        print("Files found:")
        for file in result["files"][:5]:  # Show first 5 files
            print(f"  - {file}")
    
    calc_result = mcp_demo.call_tool("calculate", expression="(10 + 5) * 2")
    if calc_result["success"]:
        print(f"\nCalculation result: {calc_result['result']}")
    
    return "MCP Demo completed successfully!"

# Execute the demo
result = await mcp_notebook_demo()
print(f"\n{result}")
```
